### 9.Blood_Pressure

```python
df['Blood Pressure'].unique()
```

```python
# Transforming blood pressure values.
def bp_category(row):
    systolic, diastolic = row['Systolic'], row['Diastolic']
    # Creating a dictionary with all possible variants.
    switcher = {
        (systolic < 90 or diastolic < 60): 'Low',
        (90 <= systolic < 120 and 60 <= diastolic < 80): 'Normal',
        (120 <= systolic < 130 and diastolic < 80): 'Elevated',
        (130 <= systolic < 140 or 80 <= diastolic < 90): 'High Blood Pressure (Stage 1)',
        (systolic >= 140 or diastolic >= 90): 'High Blood Pressure (Stage 2)',
        (systolic > 180 or diastolic > 120): 'Hypertensive Crisis',
    }
    # Defining the category.
    category = switcher.get(True, 'Check Data')
    return category

df_temp = df['Blood Pressure'].str.split('/', expand=True)
df_temp.columns = ['Systolic', 'Diastolic']
df_temp = df_temp.astype(int)

# Applying the function to each row in the DataFrame.
df['Blood Pressure'] = df_temp.apply(bp_category, axis=1)
```

## 5.Model Training

```python
y = df['Sleep Disorder']
df.drop(['Person ID','Sleep Disorder'],axis=1,inplace=True)
x=df
```

### 5.1Train Test Split

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2,stratify=y)
```

```python
model = LogisticRegression(multi_class='ovr')
model.fit(x_train,y_train)
```

## 6. Evaluation

### Testing Data Evaluation

```python
def evaluate_model(model,ind_var,act):
    pred = model.predict(ind_var)
    
    acc = accuracy_score(act,pred)
    print('Accuracy Score =',acc)
    print('**'*20)
    
    cnf_matrix = confusion_matrix(act,pred)
    print('Confusion matrix: \n',cnf_matrix)
    print('**'*20)
    
    mul_cnf_matrix = multilabel_confusion_matrix(act,pred)
    print('Multilabel Confusion matrix: \n',mul_cnf_matrix)
    print('**'*20)
    
    cls_rep =classification_report(act,pred)
    print('Classification Report: \n',cls_rep)
    
    return pred

print('Testing Data Evaluation'.center(60,'*'))

y_pred = evaluate_model(model,x_test,y_test)
```

### Training Data Evaluation

```python
print('Training Data Evaluation'.center(60,'*'))
y_pred_train = evaluate_model(model, x_train, y_train)
```
